
### Techniques and Suitability

#### 1. **Isolation Forest**
- Efficient for large-scale applications.
- Works well with high-dimensional data.

#### 2. **Interquartile Range (IQR)**
- Can be used. Good against non-normal distributions.

#### Don't use:
- **DBSCAN**: Challenging to tune for high-dimensional data.
- **Autoencoders**: Computationally expensive and complex.
- **One-Class SVM**: Computationally intensive.
- **Z-Score/Standard Deviation**: Don't use. Assumes a normal distribution, which is unlikely for weather data.

#### Approach:
I use an ensemble method between **IQR** and **Isolation Forest** to leverage the strengths of h techniques.
```

````python
from sklearn.ensemble import IsolationForest


def replace_outliers_isolation_forest(dataframe, contamination=0.05):
    # Select only numeric columns
    numeric_data = dataframe.select_dtypes(include=[np.number])
    
    # Columns to exclude from outlier detection
    columns_to_exclude = ['snowfall_binary', 'total_precipitation_sfc']
    numeric_data = numeric_data.drop(columns=columns_to_exclude, errors='ignore')

    # Fit Isolation Forest
    isolation_forest = IsolationForest(contamination=contamination, random_state=42)
    numeric_data['outlier'] = isolation_forest.fit_predict(numeric_data)

    # Identify and handle outliers
    outliers = numeric_data['outlier'] == -1  # Isolation Forest labels outliers as -1
    
    # Replace outliers with the median of the respective column
    for column in numeric_data.columns[:-1]:  # Exclude the 'outlier' column
        median_value = numeric_data.loc[~outliers, column].median()  # Median of non-outliers
        numeric_data.loc[outliers, column] = median_value

    # Drop the 'outlier' column and update the original dataframe
    numeric_data = numeric_data.drop(columns=['outlier'])
    dataframe[numeric_data.columns] = numeric_data

    return dataframe

def replace_outliers_iqr(dataframe):
    # Select only numeric columns
    numeric_data = dataframe.select_dtypes(include=[np.number])
    columns_to_exclude = ['snowfall_binary', 'total_precipitation_sfc']
    numeric_data = numeric_data.drop(columns=columns_to_exclude, errors='ignore')
    
    # Calculate Q1 (25th percentile) and Q3 (75th percentile) for each numeric column
    Q1 = numeric_data.quantile(0.25)
    Q3 = numeric_data.quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Determine the lower and upper bounds for each column
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replace outliers with lower or upper bounds
    numeric_data = numeric_data.clip(lower=lower_bound, upper=upper_bound, axis=1)

    # Update the original dataframe with replaced values
    dataframe[numeric_data.columns] = numeric_data

    return dataframe

def ensemble_outlier_replacement(dataframe, contamination=0.05):
    # Apply Isolation Forest
    isolation_cleaned = replace_outliers_isolation_forest(dataframe.copy(), contamination=contamination)

    # Apply IQR Method
    iqr_cleaned = replace_outliers_iqr(dataframe.copy())

    # Evaluate which method removed more extreme outliers based on standard deviation
    isolation_std = isolation_cleaned.select_dtypes(include=[np.number]).std().mean()
    iqr_std = iqr_cleaned.select_dtypes(include=[np.number]).std().mean()

    print(f"Isolation Forest STD: {isolation_std}, IQR STD: {iqr_std}")

    # Choose the method with the lower average standard deviation
    if isolation_std < iqr_std:
        print("Isolation Forest chosen for outlier replacement.")
        return isolation_cleaned
    else:
        print("IQR method chosen for outlier replacement.")
        return iqr_cleaned

# Apply the ensemble outlier replacement method
solar_data_cleaned = ensemble_outlier_replacement(solar_data)
solar_data = solar_data_cleaned
# Display the number of rows in the modified dataset
print(f"Total rows after outlier replacement: {solar_data.shape[0]}")
````

- The **IQR** is chosen because it helps evaluate which method removes more extreme outliers based on standard deviation.

````python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Dropout, BatchNormalization, Add
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate RMSPE
def rmspe(y_true, y_pred):
    y_true = np.array(y_true).flatten()
    y_pred = np.array(y_pred).flatten()
    epsilon = 1e-10
    return np.sqrt(np.mean(np.square((y_true - y_pred) / (y_true + epsilon))))

# Define the model function
def create_model(input_shape):
    inputs = Input(shape=(input_shape,))

    # First dense layer
    x = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(inputs)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    # Residual block 1
    res1 = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    res1 = BatchNormalization()(res1)
    res1 = Dropout(0.3)(res1)
    x = Add()([x, res1])

    # Residual block 2
    res2 = Dense(128, activation='relu', kernel_regularizer=l2(0.001))(x)
    res2 = BatchNormalization()(res2)
    res2 = Dropout(0.3)(res2)
    x = Add()([x, res2])

    # Final dense layers
    x = Dense(64, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = Dense(32, activation='relu', kernel_regularizer=l2(0.001))(x)
    outputs = Dense(1, activation='linear')(x)

    model = Model(inputs=inputs, outputs=outputs)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])
    return model

# Ensure X and y are defined
# X = np.array(...)  # Feature matrix
# y = np.array(...)  # Target vector

# Cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
train_r2_scores, test_r2_scores = [], []
train_mae_scores, test_mae_scores = [], []
train_rmse_scores, test_rmse_scores = [], []
train_rmspe_scores, test_rmspe_scores = [], []

fold = 1
for train_index, test_index in kf.split(X):
    print(f"\nFold {fold}:")
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = create_model(X_train.shape[1])

    # Callbacks
    early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=7, min_lr=1e-6)

    # Train the model
    history = model.fit(
        X_train, y_train,
        validation_split=0.2,
        epochs=100,
        batch_size=64,
        verbose=1,
        callbacks=[early_stopping, reduce_lr]
    )

    # Evaluate on training set
    y_train_pred = model.predict(X_train).flatten()
    train_r2 = r2_score(y_train, y_train_pred)
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
    train_rmspe = rmspe(y_train, y_train_pred)

    # Evaluate on test set
    y_test_pred = model.predict(X_test).flatten()
    test_r2 = r2_score(y_test, y_test_pred)
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)
    test_rmspe = rmspe(y_test, y_test_pred)

    # Append results
    train_r2_scores.append(train_r2)
    test_r2_scores.append(test_r2)
    train_mae_scores.append(train_mae)
    test_mae_scores.append(test_mae)
    train_rmse_scores.append(train_rmse)
    test_rmse_scores.append(test_rmse)
    train_rmspe_scores.append(train_rmspe)
    test_rmspe_scores.append(test_rmspe)

    print(f"Train R²: {train_r2:.4f}, Test R²: {test_r2:.4f}")
    print(f"Train MAE: {train_mae:.4f}, Test MAE: {test_mae:.4f}")
    print(f"Train RMSE: {train_rmse:.4f}, Test RMSE: {test_rmse:.4f}")
    print(f"Train RMSPE: {train_rmspe:.4f}, Test RMSPE: {test_rmspe:.4f}")

    fold += 1

# Print average metrics
print("\nAverage Metrics Across Folds:")
print(f"Train R²: {np.mean(train_r2_scores):.4f} ± {np.std(train_r2_scores):.4f}")
print(f"Test R²: {np.mean(test_r2_scores):.4f} ± {np.std(test_r2_scores):.4f}")
print(f"Train MAE: {np.mean(train_mae_scores):.4f} ± {np.std(train_mae_scores):.4f}")
print(f"Test MAE: {np.mean(test_mae_scores):.4f} ± {np.std(test_mae_scores):.4f}")
print(f"Train RMSE: {np.mean(train_rmse_scores):.4f} ± {np.std(train_rmse_scores):.4f}")
print(f"Test RMSE: {np.mean(test_rmse_scores):.4f} ± {np.std(test_rmse_scores):.4f}")
print(f"Train RMSPE: {np.mean(train_rmspe_scores):.4f} ± {np.std(train_rmspe_scores):.4f}")
print(f"Test RMSPE: {np.mean(test_rmspe_scores):.4f} ± {np.std(test_rmspe_scores):.4f}")

# Visualizing the training process of the last fold
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Over Epochs (Fold 5)')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()

# Visualize predictions vs true values (last fold)
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_test_pred, alpha=0.7, label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--r', label='Perfect Prediction')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted (Fold 5)')
plt.legend()
plt.show()

````
